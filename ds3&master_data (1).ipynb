```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('master_data.csv')
print(ds)
```

```python
for col in ds.select_dtypes(include=np.number).columns:
    ds[col].fillna(ds[col].mean(), inplace=True)
    print("Missing values after handling:")
    print(ds.isnull().sum())
```

DROPPING yr and mnth-

```python
ds = ds.drop(['yr','mnth'],axis=1)
print(ds)
```

Dropping duplicate data-

```python
ds = ds.drop_duplicates()
print(ds)
```

DROPPING Null values-

```python
ds.dropna(inplace=True)
print(ds)
```

Data visualization EDA-
1. BAR CHART

```python
plt.figure(figsize=(10,6))
sns.barplot(x='instant',y='cnt',data=ds)
plt.title('bar chart')
plt.xlabel('instant')
plt.ylabel('cnt')
plt.show()
```

MERGING Data_3 with master_data

```python
import pandas as pd
```

```python
df  = pd.read_csv('data_3.csv')
print(df)
ds = pd.read_csv('master_data.csv')
print(ds)
```

```python
Final_data = pd.merge(ds,df,on='dteday',how = 'inner')
print(Final_data)
```

Loading Final_data into csv format

```python
Final_data.to_csv('Final_data.csv')
```

TO CHECK IF Final_data has outliers,Co-relations,skewness or curtises?

To check if there are outliers?

```python
numerical_cols = Final_data.select_dtypes(include=np.number).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=Final_data[col])
    plt.title(f'Box plot of {col}')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

To check the Correlation-

```python
numerical_cols = Final_data.select_dtypes(include=np.number)
correlation_matrix = numerical_cols.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

To Check Skewness

```python
skewness = numerical_cols.skew()
print("\nSkewness:")
print(skewness)
```

To Check Kurtosis

```python
kurtosis = numerical_cols.kurtosis()
print("\nKurtosis:")
display(kurtosis)
```
